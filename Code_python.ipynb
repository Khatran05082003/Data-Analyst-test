### 1. Tìm số tiền đầu tư vào mỗi tỉnh trong năm? Tỉnh nào được đầu tư nhiều nhất?

```python
investment_by_province = dau_tu.groupby('Tỉnh')['Đầu tư'].sum()

max_investment_province = investment_by_province.idxmax()
max_investment_amount = investment_by_province.max()

print("Số tiền đầu tư vào mỗi tỉnh:\n", investment_by_province)
print(f"Tỉnh được đầu tư nhiều nhất: {max_investment_province} với số tiền {max_investment_amount}")
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=investment_by_province.index, y=investment_by_province.values)
plt.xticks(rotation=45)
plt.title("Số tiền đầu tư vào mỗi tỉnh")
plt.xlabel("Tỉnh")
plt.ylabel("Số tiền đầu tư")
plt.show()
```

### 2. Tìm Doanh thu sinh ra từ mỗi tỉnh này trong năm?

```python
revenue_by_province = doanh_thu.groupby('Tỉnh')['Doanh thu'].sum()
print("Doanh thu từ mỗi tỉnh:\n", revenue_by_province)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=revenue_by_province.index, y=revenue_by_province.values)
plt.xticks(rotation=45)
plt.title("Doanh thu từ mỗi tỉnh")
plt.xlabel("Tỉnh")
plt.ylabel("Doanh thu")
plt.show()
```

### 3. Tính số ROI của mỗi tỉnh theo từng tháng, năm

```python
merged_df = pd.merge(dau_tu, doanh_thu, on=['Tỉnh', 'Tháng', 'Năm'], how='left', suffixes=('_dau_tu', '_doanh_thu'))

merged_df['ROI'] = (merged_df['Doanh thu'] - merged_df['Đầu tư']) / merged_df['Đầu tư']


roi_by_province_month = merged_df.groupby(['Tỉnh', 'Tháng'])['ROI'].mean()
roi_by_province_year = merged_df.groupby('Tỉnh')['ROI'].mean()

print("ROI của mỗi tỉnh theo từng tháng:\n", roi_by_province_month)
print("ROI của mỗi tỉnh theo năm:\n", roi_by_province_year)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=roi_by_province_year.index, y=roi_by_province_year.values)
plt.xticks(rotation=45)
plt.title("ROI của mỗi tỉnh theo năm")
plt.xlabel("Tỉnh")
plt.ylabel("ROI")
plt.show()
```

### 4. Số tiền đầu tư vào mỗi Khách hàng cấp 1 và doanh thu được tạo ra trong năm?

```python
investment_by_kh_c1 = dau_tu.groupby('Khách Hàng C1')['Đầu tư'].sum()
revenue_by_kh_c1 = doanh_thu.groupby('Khách hàng C1')['Doanh thu'].sum()

print("Số tiền đầu tư vào mỗi Khách hàng cấp 1:\n", investment_by_kh_c1)
print("Doanh thu từ mỗi Khách hàng cấp 1:\n", revenue_by_kh_c1)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=investment_by_kh_c1.index, y=investment_by_kh_c1.values, color='blue', label='Đầu tư')
sns.barplot(x=revenue_by_kh_c1.index, y=revenue_by_kh_c1.values, color='red', alpha=0.6, label='Doanh thu')
plt.xticks(rotation=45)
plt.title("Số tiền đầu tư và doanh thu của mỗi Khách hàng cấp 1")
plt.xlabel("Khách hàng cấp 1")
plt.ylabel("Giá trị")
plt.legend()
plt.show()
```

### 5. Tìm Số lượng nông dân trong hệ thống của mỗi Khách hàng cấp 2, khách hàng nào có số nông dân cao nhất?

```python
farmers_by_kh_c2 = dau_tu.groupby('Khách hàng C2')['SL Nông Dân'].sum()

max_farmers_kh_c2 = farmers_by_kh_c2.idxmax()
max_farmers_amount = farmers_by_kh_c2.max()

print("Số lượng nông dân trong hệ thống của mỗi Khách hàng cấp 2:\n", farmers_by_kh_c2)
print(f"Khách hàng cấp 2 có số nông dân cao nhất: {max_farmers_kh_c2} với số lượng {max_farmers_amount}")
```

```python
top_20_farmers_by_kh_c2 = farmers_by_kh_c2.sort_values(ascending=False).head(20)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_20_farmers_by_kh_c2.index, y=top_20_farmers_by_kh_c2.values)
plt.xticks(rotation=45)
plt.title("Số lượng nông dân của mỗi Khách hàng cấp 2 (Top 20)")
plt.xlabel("Khách hàng cấp 2")
plt.ylabel("Số lượng nông dân")
plt.show()
```

### 6. Có bao nhiêu khách hàng mà công ty đầu tư nhưng không tạo ra doanh thu trong năm?

```python
dau_tu.rename(columns={"Khách Hàng C1": "Khách hàng C1"}, inplace=True)
```

```python
investment_revenue_merged = pd.merge(dau_tu, doanh_thu, on='Khách hàng C1', how='left', suffixes=('_dau_tu', '_doanh_thu'))

customers_no_revenue = investment_revenue_merged[(investment_revenue_merged['Đầu tư'] > 0) & (investment_revenue_merged['Doanh thu'].isna())]['Khách hàng C1'].unique()

print(f"Số lượng khách hàng mà công ty đầu tư nhưng không tạo ra doanh thu trong năm: {len(customers_no_revenue)}")
print("Danh sách các khách hàng này:", customers_no_revenue)
```

### 9. So sánh tổng mức đầu tư giữa hai nhóm khách hàng

```python
investment_per_cust1 = dau_tu.groupby('Khách hàng C1')['Đầu tư'].sum().reset_index()

investment_per_cust2 = dau_tu.groupby('Khách hàng C2')['Đầu tư'].sum().reset_index()

print("Tổng mức đầu tư theo khách hàng cấp 1:")
print(investment_per_cust1)

print("\nTổng mức đầu tư theo khách hàng cấp 2:")
print(investment_per_cust2)
```

```python
plt.figure(figsize=(12, 6))
investment_per_cust1['Loại khách hàng'] = 'Khách hàng Cấp 1'
investment_per_cust2['Loại khách hàng'] = 'Khách hàng Cấp 2'
investment_combined = pd.concat([investment_per_cust1, investment_per_cust2.rename(columns={'Khách hàng C2': 'Khách hàng C1'})])

sns.boxplot(x='Loại khách hàng', y='Đầu tư', data=investment_combined)
plt.title('So sánh mức đầu tư giữa Khách hàng Cấp 1 và Cấp 2')
plt.xlabel('Loại khách hàng')
plt.ylabel('Mức đầu tư')
plt.show()
```

### 10. Áp dụng mô hình kmeans phân loại phân khúc khách hàng

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler


data_cust1 = dau_tu.groupby('Khách hàng C1')['Đầu tư'].sum().reset_index()
data_cust2 = dau_tu.groupby('Khách hàng C2')['Đầu tư'].sum().reset_index()


scaler = StandardScaler()
data_cust1_scaled = scaler.fit_transform(data_cust1[['Đầu tư']])
data_cust2_scaled = scaler.fit_transform(data_cust2[['Đầu tư']])


kmeans_cust1 = KMeans(n_clusters=3, random_state=42)
kmeans_cust1.fit(data_cust1_scaled)
data_cust1['Cluster'] = kmeans_cust1.labels_

kmeans_cust2 = KMeans(n_clusters=3, random_state=42)
kmeans_cust2.fit(data_cust2_scaled)
data_cust2['Cluster'] = kmeans_cust2.labels_

print(data_cust1)
print(data_cust2)


plt.figure(figsize=(10, 6))
sns.scatterplot(x='Khách hàng C1', y='Đầu tư', hue='Cluster', data=data_cust1, palette='viridis')
plt.title('Customer Segmentation for Khách hàng C1')
plt.show()

plt.figure(figsize=(10, 6))
sns.scatterplot(x='Khách hàng C2', y='Đầu tư', hue='Cluster', data=data_cust2, palette='viridis')
plt.title('Customer Segmentation for Khách hàng C2')
plt.show()
```
